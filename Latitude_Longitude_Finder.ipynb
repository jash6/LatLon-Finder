```python
df= pd.read_csv('Final.csv')
df
```

```python
df_temp = df.loc[0:29,['State','Total']]
df_temp
```

```python
list_lat = [] 
list_long = []
for index, row in df_temp.iterrows(): 
    State = row['State']       
    query = str(State)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

	
# create new columns from lists    

df_temp['lat'] = list_lat   

df_temp['lon'] = list_long
```

```python
df_temp
```

```python
df_final= df_temp.set_index('State')
df_final
```
