Imports

```python
import pandas as pd
import numpy as np
from opencage.geocoder import OpenCageGeocode
```

```python
Reading CSV
```

```python
df= pd.read_csv('Pan_india_shop_info.csv')
df
```

```python
df_temp= df[['STATE NAME','DISTRICT NAME','CITY NAME']]
df_temp
```

Groupby District Name 

```python
df_temp['Counts'] = np.zeros(len(df))
grp_df = df_temp.groupby(['DISTRICT NAME']).count()
grp_df.reset_index(inplace=True)
grp_df
```

Get The LAT LON from opencage and add it to the Data frame

```python
from opencage.geocoder import OpenCageGeocode
key = 'Your Key'
geocoder = OpenCageGeocode(key)
list_lat = [] 
list_long = []
for index, row in grp_df.iterrows(): 
    District = row['DISTRICT NAME']       
    query = str(District)
    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']
    list_lat.append(lat)
    list_long.append(long)

	
# create new columns from lists    

grp_df['lat'] = list_lat   

grp_df['lon'] = list_long
```

```python
grp_final=grp_df[['DISTRICT NAME','Counts','lat','lon']]
grp_final
```

```python
df_final= grp_final.set_index('DISTRICT NAME')
df_final
```
